```python
# Import Dependencies
import pandas as pd
import os
```

```python
# Get the file path for running team one
file1=os.path.join('../','Resources', 'team-1.csv')
file2=os.path.join('../','Resources', 'team-2.csv')
```

```python
# Read the CSV file into a DataFrame
df=pd.read_csv(file1)
df.head()
```

```python
df2=pd.read_csv(file2)
df2.head()
```

```python
#joining the two dataframes
Frames=[df,df2]
final_df=pd.concat(Frames,)
final_df()
```

```python
# Melt the DataFrame
final_df.melt()
```

```python
# Melt the DataFrame using the variable we'd like to keep in the long DF
meltrunner_df=final_df.melt(id_vars="runner_name")
meltrunner_df
```

```python
# Melt the DataFrame using the variable we'd like to keep in the long DF
meltedrunner_df=final_df.melt(id_vars=["runner_name"], value_vars= ['coach_name', 'run_1', 'run_2','run_3'])
meltedrunner_df
```

```python
# Renaming the columns
#meltedrunner_df_renamed = pd.melt(df, id_vars="runner_name", var_name="coach/runID", value_name="time")
#meltedrunner_df_renamed
#Using .rename(columns={}) in order to rename columns
renamed_df = meltedrunner_df.rename(columns={"runner_name":"Athlete", "variable":"Run ID", "value":"Time"})
renamed_df
```

```python
# Remove coach name from DataFrame
Cleaned_df = renamed_df.drop(labels=range(0, 7), axis=0)
```

```python
# Show final result
Cleaned_df
```

```python
#set indices
final_clean_df=Cleaned_df.set_index('Athlete')
final_clean_df
```

```python
# Save results to a file
final_clean_df.to_csv('final_clean_out.csv', index=False)  
```
